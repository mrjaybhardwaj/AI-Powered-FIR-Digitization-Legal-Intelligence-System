```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with fitz.open(pdf_path) as pdf_doc:
            for page_num in range(pdf_doc.page_count):
                page = pdf_doc[page_num]
                text += page.get_text()
    except Exception as e:
        print(f"Error extracting text from PDF: {e}")
    
    return text

# Example usage
pdf_path = "d:/Project/FIR/FIR COPY/FIR 2.pdf"
provided_text = extract_text_from_pdf(pdf_path)
provided_text
```

```python
import re

def clean_text(input_text):
    # Remove newline characters and extra whitespaces
    cleaned_text = re.sub(r'\n+', ' ', input_text)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)

    # Remove unnecessary characters and patterns
    cleaned_text = re.sub(r'[^\w\s\d/:.-]', '', cleaned_text)

    return cleaned_text.strip()

# Example usage:

cleaned_text = clean_text(provided_text)

print(cleaned_text)
```

```python
import re

def extract_information(fir_text):
    # Extract FIR number, location, and police station directly
    direct_extraction_match = re.search(r'Cr\.P\.C\. (\d+).*? (\w+) (\d+) (\w+)', fir_text)
    year, police_station, fir_number, location = direct_extraction_match.groups() if direct_extraction_match else (None, None, None, None)

    # Extract incident date and time value
    # Extract incident date and time value using the specified pattern
    incident_datetime_match = re.search(r'(\d{2}/\d{2}/\d{4} \d{2}:\d{2}:\d{2})', fir_text)
    incident_datetime = incident_datetime_match.group(1) if incident_datetime_match else None
    # Extract crime description from the longest text with "/"
    crime_description_match = max(re.finditer(r'[^/]+', fir_text), key=lambda match: len(match.group(0)))
    crime_description = crime_description_match.group(0).strip() if crime_description_match else None

    # Display extracted information
    print(f"Year: {year}")
    print(f"Police Station: {police_station}")
    print(f"FIR Number: {fir_number}")
    print(f"Location: {location}")
    print(f"Incident Date and Time: {incident_datetime}")
    print(f"Crime Description: {crime_description}")

# Sample usage with the provided FIR report
extract_information(cleaned_text)
```

```python
from googletrans import Translator

def split_and_translate(text, max_length=500):
    # Split the text into smaller parts based on the specified max length
    chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]
    
    # Initialize Google Translator
    translator = Translator()

    # Translate each chunk and accumulate the results
    translated_text = ""
    for chunk in chunks:
        translation = translator.translate(chunk, src='auto', dest='en')
        translated_text += translation.text + " "

    return translated_text.strip()

# Example usage:


translated_result = split_and_translate(hindi_text)
print(translated_result)
```

```python
import re

def extract_dates(text):
    date_patterns = [re.compile(r'\d{2}/\d{2}/\d{4}'), re.compile(r'\d{2}/\d{2}/\d{2}')]
    dates = []
    for pattern in date_patterns:
        matches = pattern.findall(text)
        dates.extend(matches)
    return dates

def extract_fir_details(text):
    # Regular expression patterns for extracting information
    section_pattern = re.compile(r'धारा (\d+)')
    location_pattern = re.compile(r'थाना (\w+)')
    date_pattern = [re.compile(r'\d{2}/\d{2}/\d{4}'), re.compile(r'\d{2}/\d{2}/\d{2}')]

    # Additional patterns for complainant information
    name_pattern = re.compile(r'नाम: (.+?)\n')
    father_name_pattern = re.compile(r'पिता का नाम: (.+?)\n')
    dob_pattern = re.compile(r'जन्म तिथि/ वर्ष: (.+?)\n')

    # Pattern for capturing everything after the word "सेवामे"
    description_pattern = re.compile(r'सेवामे([\s\S]*)')

    # Pattern for capturing everything after the words "Action taken" until "."
    action_taken_pattern = re.compile(r'Action taken :[\s\S]*?(\..*?)(?=\n|$)')

    # Extracting information using regular expressions
    section_match = section_pattern.search(text)
    location_match = location_pattern.search(text)
    
    # Extracting dates using the previously defined function
    dates = extract_dates(text)
    
    # Joining extracted dates into a single string
    date = ', '.join(dates)
    
    description_match = description_pattern.search(text)
    
    # Extracting complainant information
    name_match = name_pattern.search(text)
    father_name_match = father_name_pattern.search(text)
    dob_match = dob_pattern.search(text)
    
    # Extracting action taken information
    action_taken_match = action_taken_pattern.search(text)

    # Getting matched groups
    section = section_match.group(1) if section_match else None
    location = location_match.group(1) if location_match else None
    description = description_match.group(1).strip() if description_match else None
    name = name_match.group(1) if name_match else None
    father_name = father_name_match.group(1) if father_name_match else None
    dob = dob_match.group(1) if dob_match else None
    action_taken = action_taken_match.group(1) if action_taken_match else None

    return {
        "section": section,
        "location": location,
        "date": date,
        "description": description,
        "name": name,
        "father_name": father_name,
        "dob": dob,
        "action_taken": action_taken
    }



fir_details = extract_fir_details(hindi_text)
print(fir_details)
```
